"""Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Answer:- The url_for() function is used to build a url to the specific function dynamicllay. The first argument is the name of the specified function, and then we can pass any number of keyword argument corresponding to the variable part of the url"""

from flask import Flask,redirect,url_for
from flask import request
app=Flask(__name__)
@app.route("/success")
def success():
    return "you are pass you get a success"
@app.route("/fail")
def fail():
    return "you are fail in the exam"
@app.route("/results")
def results():
    marks=int(request.args.get('x'))
    if marks >=45:
        return redirect(url_for('success'))
    else:
        return redirect(url_for('fail'))
if __name__=='__main__':
    app.run(host="0.0.0.0", port =8080)
